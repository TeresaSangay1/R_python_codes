## Regular expresion R 

```r
library("readxl")
library("stringr") # library to RE
library("dplyr") # grepl para detertar patrones en texto 
```

```r
junin_data = read_excel("../../data/Region_Junin.xlsx")
```

```r
junin_data
```

```r
attach(junin_data) #para que cada variable sea un objeto
```

### 1.0 select districs that contanis "AC"

```r
newbase <- dplyr::filter(junin_data, grepl('AC', District))
```

```r
head(newbase)
```

```r
dplyr::filter(junin_data, grepl('pacha', Place))
```

```r
junin_data %>% filter(grepl('pacha', Place, ignore.case=TRUE))
```

```r
str_view(junin_data$District, regex("CIUDAD", ignore_case = TRUE))
```

```r
junin_data %>% filter(grepl('CIUDAD', District, ignore.case=TRUE))
```

```r
junin_data %>% filter(grepl('yauli', District, ignore.case=TRUE))
```

```r
junin_data %>% filter(grepl('del', Place, ignore.case=TRUE))
```

```r
# # Beginning word with hu

junin_data %>% filter(grepl('^hu', District, ignore.case=TRUE))
```

```r
junin_data %>% filter(grepl('^huari', District, ignore.case=TRUE))
```

```r
# Ending word
junin_data %>% filter(grepl('ro$', Place, ignore.case=TRUE))
```

```r
junin_data %>% filter(grepl('ca$', Place, ignore.case=TRUE))
```

```r
junin_data %>% filter(grepl('am*', Place, ignore.case=TRUE))

# match : a , m o am
```

```r
junin_data %>% filter(grepl('am+', Place, ignore.case=TRUE))

# match : am (strict)
```

```r
junin_data %>% filter(grepl('am?', Place, ignore.case=TRUE))

# match a or am
```
