## 2 <a id='2'> Check duplicates in a group of columns </a>  

```r
enapres2020_1 %>% group_by(CCDD ,CCPP , CCDI ,CONGLOMERADO , NSELV, VIVIENDA, HOGAR) %>% 
                  mutate(duplicates = n()) %>% filter(duplicates >1) %>%
                select(CCDD ,CCPP , CCDI ,CONGLOMERADO , NSELV, VIVIENDA, HOGAR) 

# First group by unique household identifier
# mutate() create variables
# filter duplicates 
# select unique household identifier
```

##### Function distinct() [dplyr package] can be used to keep only unique/distinct rows from a data frame. 

```r
# Drop duplicates rows (observaciones)

enapres2020_1_ndup <- enapres2020_1 %>% distinct(CCDD ,CCPP , CCDI ,CONGLOMERADO , NSELV, VIVIENDA, HOGAR, .keep_all = TRUE)
```

```r
head(enapres2020_1_ndup)
```

```r
# Check unique values. We can se that there is missing values

unique(ESTRATO)
```

```r
enapres2020_1_ndup <- enapres2020_1_ndup %>%  mutate(Dummy_2 = ifelse(ESTRATO == 4 ,  1 , ifelse(!is.na(ESTRATO),0, NA) ) )
```

```r
enapres2020_1_ndup <- enapres2020_1_ndup %>%  mutate(Dummy_3 = ifelse( ESTRATO %in% c(3,4,5) ,  1 , ifelse(!is.na(ESTRATO),0, NA) ) )
```

```r
enapres2020_1_ndup 
```

```r
# There are NA because we use all dataset (complete and incomplete household survey)

df <- enapres2020_1_ndup %>% group_by(CCDD ,CCPP , CCDI) %>% 
                    summarise(estrato_D_median = median(!is.na(Dummy_2)), estrato_D_mean = mean(!is.na(Dummy_2)),
                    estrato_D_max = max(!is.na(Dummy_2)), estrato_D_min = min(!is.na(Dummy_2)),
                    estrato_D_total =sum(!is.na(Dummy_2)), total_hogares_distrito = n() )


# !is.na(Dummy_2) not take nan values
```
