```python
# load Domestic Violence module

rech84 = pd.read_stata('../../data/Endes/rec84dv.dta') 

```

```python
rech84.info
```

```python
rech84['caseid']
```

```python
rech84['caseid'] = rech84['caseid'].str.strip()  # drop empty spaces 
```

```python
rech84['caseid']
```

```python
rech84['caseid'] = rech84['caseid'].str[:9] + rech84['caseid'].str[9:12].str.strip()


```

```python
rech84
```

```python
rech84.columns
```

```python
# handle missing values 

index = np.where( rech84.columns.str.contains('d101'))[0]

names = list(rech84.columns[index])
```

```python
for i in names:
    rech84[f'{i}'] = rech84[f'{i}'].replace(9,np.nan)
    rech84[f'{i}'] = rech84[f'{i}'].replace(99,np.nan)

# replace 9 or 99 to missing value
```
